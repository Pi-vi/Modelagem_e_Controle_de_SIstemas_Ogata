```python
from control import *
from numpy import linspace
import matplotlib.pyplot as plt
```

Considere o sistema:

$\left[ \begin{array}{c} \dot{x}_1 \\ \dot{x}_2 \end{array} \right] =\left[ \begin{array}{c} -1 & 0,5 \\ -1 & 0 \end{array} \right] \left[ \begin{array}{c} x_1 \\ x_2 \end{array} \right] + \left[ \begin{array}{c} 0 \\ 1 \end{array} \right] u$

$y = \left[ \begin{array}{c}1 & 0 \end{array} \right] \left[ \begin{array}{c} x_1 \\ x_2 \end{array} \right]$

Obtenha as curvas de resposta $y(t)$ quando a entrada $u$ é dada por: 

1. $u$ = entrada em degrau unitário
2. $u = e^{-t}$

Suponha que o estado inicial $\vec{x}(0) = \vec{0}$.

```python
t = linspace(0,12,1001)
```

```python
A = [[-1,0.5],[-1,0]]
B = [[0],[1]]
C = [1,0]
D = [0]
```

```python
sistema = ss(A,B,C,D)
sistema
```

Reposta ao degrau unitário

```python
t1,y1 = step_response(sistema,t)
```

```python
plt.figure(dpi = 150)
plt.plot(t,y1)
plt.xlabel('Tempo(s)')
plt.ylabel('Saída')
plt.title('Resposta ao degrau unitário')
plt.grid(True)
plt.show()
```

Para a resposta à entrada exponencial:

```python
from numpy import exp
```

```python
u = exp(-t)
```

```python
t2,y2 = forced_response(sistema,t,u)
```

```python
plt.figure(dpi = 150)
plt.plot(t,u)
plt.plot(t,y2)
plt.legend(['Entrada exponencial','Saída'])
plt.xlabel('Tempo(s)')
plt.ylabel('Saída')
plt.title(r'Resposta à entrada exponencial $u = e^{-t}$')
plt.grid(True)
plt.show()
```
