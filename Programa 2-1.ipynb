```python
from control import *
```

Criando a função de transferência:

```python
num1 = [10]
den1 = [1,2,10]
num2 = [5]
den2 = [1,5]
```

Função de transferência $G_1(s)$:

```python
G1 = tf(num1,den1)
G1
```

Função de transferência $G_2(s)$:

```python
G2 = tf(num2,den2)
G2
```

## Letra (a):

```python
Ga = series(G1,G2)
Ga
```

## Letra (b):

```python
Gb = parallel(G1,G2)
Gb
```

## Letra (c):

```python
Gc = feedback(G1,G2)
Gc
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,20,1000)
t1,y1 = impulse_response(Gc,t)
```

```python
plt.figure(dpi=150)
plt.plot(t1,y1)
plt.xlabel('tempo[s]')
plt.ylabel('Amplitude')
plt.grid(True)
```
